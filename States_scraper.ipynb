```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
import pprint
```

```python
states_csv = pd.read_csv("../Population_metrics_by_state_D3_visualizations/assets/data/data.csv")
states_csv = states_csv.drop(8).reset_index()
# states_values = pd.DataFrame(states_csv["state"])

states_values = states_csv["state"]
states_values
```

```python
images_urls = []
count = 1
for x in states_values:
#     print(count)
    val = count % 2
#     print(f"val is equal to: {val}")
    if ( val == 0):
#         print("llego al false")
        images_urls.append({"state":x, "image":"", "success":False})
#         rows['success'] = False
        
    else:
#         images_urls.append({"state":x, "image":"", "success":True})
        images_urls.append({"state":x, "image":"", "success":False})
#         rows['success'] = True
    count = count + 1
    
#     print(x)
    
# images_urls[0]['success']
states_values['image'] = ""
print(images_urls)
# print(states_values)
```

```python
def checkflag(diction):
    count = 0
    for x in diction:
        if (x['success']):
            
            success= True
            
        else :
            success = False
            count+=1
#             return success
    print(f"you have {count} of falses")
    return success

checkflag(images_urls)
```

```python
for x in range(len(images_urls)):
    
    images_urls[x]['image']="https://geology.com/state-map/maps/"+images_urls[x]['state'].lower().replace(" ","-")+"-county-map.gif"
    images_urls[x]['success']=True
    
pprint.pprint(images_urls)
import json

with open('states_images.json', 'w') as fp:
    json.dump(images_urls, fp)
```

```python
# this one is the good one 

flag = False

url = 'https://geology.com/world/the-united-states-of-america-satellite-image.shtml'
while flag == False:
    
    try:
        executable_path = {'executable_path': 'chromedriver.exe'}
        browser = Browser('chrome', **executable_path, headless=False)
        browser.visit(url)
    
        for x in range(len(images_urls)):
            print(images_urls[x]['success'])
            if (images_urls[x]['success']==False):
                print("it got into the if")
                print(f"we are looking for the state {images_urls[x]['state']}")
                time.sleep(5)
                one_state = browser.find_by_text(images_urls[x]["state"]).click()
                
                html = browser.html
                soup = bs(html,'html.parser')
                image = soup.select("article a img")[0]['src']
                
                images_urls[x]['image']="https://geology.com/state-map/"+image
                images_urls[x]['success']=True
            #     time.sleep(5)
                print(f"we changed {images_urls[x]}")
                browser.back()
            else:
                print('next state')
                print(f"we skiped {images_urls[x]}")

                print(x)
                x+=1
        browser.quit()
        
    except:
        print("error! leaving browser")
        browser.quit()
        time.sleep(10)
        
        
    
    
    flag = checkflag(images_urls)
    print("end of while loop")
    print('-----------------------------')
    print('****************************')
    print(f'Flag value = {flag}')
    print('-----------------------------')
    print('****************************')
    
    
    
```
